```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hotel_bookings_cleaned.csv")
```

```python
 df.isnull().sum()
```

```python
# Booking Cancellation Distribution

df['is_canceled'].value_counts().plot(kind='bar')
plt.xlabel("Canceled (0 = No, 1 = Yes)")
plt.ylabel("Number of Bookings")
plt.title("Booking Cancellation Distribution")
plt.show()
```

```python
# Average Daily Rate (ADR) by Hotel Type

df.groupby('hotel')['adr'].mean().plot(kind='bar')
plt.xlabel("Hotel Type")
plt.ylabel("Average Daily Rate (ADR)")
plt.title("Average ADR by Hotel Type")
plt.show()
```

```python
# Monthly Booking Trend

df['arrival_date_month'].value_counts().plot(kind='line')
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.title("Monthly Booking Trend")
plt.show()
```

```python
# Market Segment Distribution

df['market_segment'].value_counts().plot(kind='bar')
plt.xlabel("Market Segment")
plt.ylabel("Number of Bookings")
plt.title("Market Segment Distribution")
plt.show()
```

```python
# ADR vs Lead Time

plt.scatter(df['lead_time'], df['adr'])
plt.xlabel("Lead Time (days)")
plt.ylabel("ADR")
plt.title("ADR vs Lead Time")
plt.show()
```

```python
# Top 10 Countries by Bookings

df['country'].value_counts().head(10).plot(kind='bar')
plt.xlabel("Country")
plt.ylabel("Bookings")
plt.title("Top 10 Countries by Number of Bookings")
plt.show()
```

```python
# Guest Composition Analysis

guest_data = df[['adults', 'children', 'babies']].sum()

guest_data.plot(kind='bar')
plt.xlabel("Guest Type")
plt.ylabel("Count")
plt.title("Guest Composition Distribution")
plt.show()
```
